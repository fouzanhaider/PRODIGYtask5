```python
# Install scapy if not already installed
!pip install scapy

# Import necessary libraries
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output
import threading
import pandas as pd
from scapy.all import sniff, IP, TCP, UDP, Raw

# Function to process packets
def process_packet(packet):
    if packet.haslayer(IP):
        ip_layer = packet.getlayer(IP)
        src_ip = ip_layer.src
        dst_ip = ip_layer.dst
        protocol = packet.proto
        payload = packet.load if packet.haslayer(Raw) else "No Payload"

        packet_info = {
            "Source IP": src_ip,
            "Destination IP": dst_ip,
            "Protocol": protocol,
            "Payload": payload
        }
        packets.append(packet_info)
        update_table()

# Function to update the display table
def update_table():
    clear_output(wait=True)
    df = pd.DataFrame(packets)
    display(df)

# Initialize packets list
packets = []

# Widgets for user input
start_button = widgets.Button(description="Start Sniffing")
stop_button = widgets.Button(description="Stop Sniffing")

# Define sniff_thread as a global variable
sniff_thread = None

# Function to handle button clicks
def start_sniffing(_):
    global sniff_thread
    sniff_thread = threading.Thread(target=lambda: sniff(prn=process_packet, store=0))
    sniff_thread.start()

def stop_sniffing(_):
    global sniff_thread
    if sniff_thread is not None:
        sniff_thread.do_run = False

start_button.on_click(start_sniffing)
stop_button.on_click(stop_sniffing)

# Display widgets
display(start_button, stop_button)
```
